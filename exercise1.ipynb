```python
# importing packages
import numpy as np
import pandas as pd
import matplotlib as plt

%matplotlib inline
```

```python
# read the source csv
url = 'https://raw.githubusercontent.com/ine-rmotr-curriculum/FreeCodeCamp-Pandas-Real-Life-Example/master/data/sales_data.csv'
sales = pd.read_csv(url, parse_dates=['Date'])
```

```python
# Customer age mean
sales['Customer_Age'].mean()
```

```python
# show box chart Customer_Age
sales['Customer_Age'].plot(kind='box', vert=False, figsize=(14, 6))
```

```python
# show density chart Customer_Age
sales['Customer_Age'].plot(kind='density', figsize=(14, 6))
```

```python
# Show Order_Quantuity mean
sales['Order_Quantity'].mean()
```

```python
# show Order Quantity chat with histogram
sales['Order_Quantity'].plot(kind='hist', bins=30, figsize=(14,6))
```

```python
# Show Order Quantity chart with box chart
sales['Order_Quantity'].plot(kind='box', vert=False, figsize=(14, 6))
```

```python
# count sales per Year
sales['Year'].value_counts()
```

```python
# Show sales per Year with pie chart
sales['Year'].value_counts().plot(kind='pie', figsize=(6, 6))
```

```python
# show sales per Month
sales['Month'].value_counts()
```

```python
# show sales per Month with bar chart
sales['Month'].value_counts().plot(kind='bar',  figsize=(14, 6))
```

```python
# show the country has the most sales Quantity
sales['Country'].value_counts().head(1)
```

```python
# show all country sales
sales['Country'].value_counts()
```

```python
# show country sales with bar chart
sales['Country'].value_counts().plot(kind='bar', figsize=(14, 6))
```

```python
# make a list the product that sold
sales.loc[:, 'Product'].unique()
```

```python
# show 10 most sales product
sales['Product'].value_counts().head(10).plot(kind='bar', figsize=(14, 6))
```

```python
# shows the relationship between Unit_Cost and Unit_Price with scatter chart
sales.plot(kind='scatter', x='Unit_Cost', y='Unit_Price', figsize=(6, 6))
```

```python
# shows the relationship between Order Quantity and Profit with scatter chart 
sales.plot(kind='scatter', x='Order_Quantity', y='Profit', figsize=(6, 6))
```

```python
# show the relationship between country and profit with boxplot
sales[['Profit', 'Country']].boxplot(by='Country', figsize=(10, 6))
```

```python
# show the relationship between Customer_Age and Country with box_plot
sales[['Customer_Age', 'Country']].boxplot(by='Country', figsize=(10,6))
```

```python
# calculate new calculate date form
sales['Calculated_Date'] = sales[['Year', 'Month', 'Day']].apply(lambda x: '{} {} {}'.format(x[0], x[1], x[2]), axis=1)

sales['Calculated_Date'].head()
```

```python
# parse from calculate the date to DateTime object
sales['Calculated_Date'] = pd.to_datetime(sales['Calculated_Date'], errors='coerce')

sales['Calculated_Date'].head()
```

```python
# show how much sales evolve through the years
sales['Calculated_Date'].value_counts().plot(kind='line', figsize=(10,6))
```

```python
# show revenue before +50
sales['Revenue']
```

```python
# add 50 to all revenue
sales['Revenue'] += 50
```

```python
# show revenue after adding it by 50
sales['Revenue']
```

```python
# show how many orders were made in Canada or France
sales.loc[(sales['Country'] == 'Canada') | (sales['Country'] == 'France')].shape[0]
```

```python
# show how many bike racks were made in Canada
sales.loc[(sales['Country'] == 'Canada') & (sales['Sub_Category'] == 'Bike Racks')].shape[0]
```

```python
# show how many order in each region of france
france_states = sales.loc[sales['Country'] == 'France', 'State'].value_counts()
france_states
```

```python
# show how many order in each region of france with bar chart
france_states.plot(kind='bar', figsize=(10,6))
```

```python
# show how many sales per category
sales['Product_Category'].value_counts()
```

```python
# show how many sales per category with pie chart
sales['Product_Category'].value_counts().plot(kind='pie', figsize=(6, 6))
```

```python
# see how many orders made a per accessori sub-categories
accessories = sales.loc[(sales['Product_Category'] == 'Accessories', 'Sub_Category')].value_counts()
accessories
```

```python
# see how many orders made a per accessori sub-categories in bar chart
accessories.plot(kind='bar', figsize=(14,6))
```

```python
# show how many orders were made per bike Sub Category with pie chart
bike.plot(kind='pie', figsize=(6,6))
```

```python
# show what gender has the most amount sales
sales['Customer_Gender'].value_counts()
```

```python
# show what gender has the most amount sales with bar chart
sales['Customer_Gender'].value_counts().plot(kind='bar', figsize=(6,6))
```

```python
# show how many sales more than 500 by Men
sales.loc[(sales['Customer_Gender'] == 'M') & (sales['Revenue'] == 500)].shape[0]
```

```python
# show top 5 highest revenue
sales.sort_values(['Revenue'], ascending=False).head(5)
```
